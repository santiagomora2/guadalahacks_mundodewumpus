```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pd_speech_features.csv')
df.head()
```

```python
df.shape
```

```python
from sklearn.utils import resample

df_majority = df[df.iloc[:,754]==1]
df_minority = df[df.iloc[:,754]==0]
 
# Downsample majority class
df_majority_downsampled = resample(df_majority, 
                                 replace=False,    
                                 n_samples=130)
#Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     
                                 n_samples=190)
# Combine minority class with downsampled majority class
df_up_down_sampled = pd.concat([df_majority_downsampled, df_minority_upsampled])
```

```python
df['locAbsJitter']
```

```python
# MFCC: Col_57 to Col_140
df_MFCC = df_up_down_sampled.iloc[:, 56:140]
df_MFCC.drop(columns = ['mean_Log_energy', 'mean_delta_log_energy', 'mean_delta_delta_log_energy', 'std_Log_energy', 'std_delta_log_energy', 'std_delta_delta_log_energy'], inplace = True)
df_MFCC_ = df_MFCC.iloc[:, 5:]
df_MFCC_.head()
```

```python
df_idgen = df_up_down_sampled.iloc[:, 0:2]
df_idgen.head()
```

```python
df_class = df_up_down_sampled.loc[:, 'class']
df_class.head()
```

```python
df_ = pd.concat([df_MFCC_, df_class], axis = 1)
df_.head()
```

```python
X = df_.iloc[:, :-1]
y = df_.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1234, stratify = y)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(X_train, y_train);
```

```python
y_pred = LR.predict(X_test)

accuracy = (y_test == y_pred).mean()
accuracy
```

```python
from sklearn.model_selection import cross_validate;

cv_8 = cross_validate(estimator = LR,
                      cv = 8,
                      X = X_train,
                      y = y_train);
```

```python
res = pd.DataFrame(cv_8)
res.agg(["mean", "sem"])
```

# SVM

```python
from sklearn.svm import SVC


SVMC = SVC(kernel = 'sigmoid')
SVMC.fit(X_train, y_train);
```

```python
y_pred = SVMC.predict(X_test)
accuracy = (y_test == y_pred).mean()
accuracy
```

```python
cv_8 = cross_validate(estimator = SVMC,
                      cv = 8,
                      X = X_train,
                      y = y_train);
```

```python
res = pd.DataFrame(cv_8)
res.agg(["mean", "sem"])
```

# SVMC (otro kernel) 

```python

SVMC1 = SVC(kernel = 'linear')
SVMC1.fit(X_train, y_train);
```

```python
cv_8 = cross_validate(estimator = SVMC1,
                      cv = 8,
                      X = X_train,
                      y = y_train);
```

```python
res = pd.DataFrame(cv_8)
res.agg(["mean", "sem"])
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNNC = KNeighborsClassifier()
KNNC.fit(X_train, y_train);
```

```python
y_pred = KNNC.predict(X_test)
accuracy = (y_test == y_pred).mean()
accuracy
```

```python
cv_8 = cross_validate(estimator = KNNC,
                      cv = 8,
                      X = X_train,
                      y = y_train);
```

```python
res = pd.DataFrame(cv_8)
res.agg(["mean", "sem"])
```

# arbOles

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(random_state = 0)
RFC.fit(X_train, y_train);
```

```python
y_pred = RFC.predict(X_test)
accuracy = (y_test == y_pred).mean()
accuracy
```

```python
cv_8 = cross_validate(estimator = RFC,
                      cv = 8,
                      X = X_train,
                      y = y_train);
res = pd.DataFrame(cv_8)
res.agg(["mean", "sem"])
```

```python
#modelos = LR, SVMC, KNNC, RFC, SVMC1
import joblib
```

```python
ref_cols = list(X.columns)
target = 'class'
```

```python
joblib.dump(value = [LR, ref_cols, target], filename = 'LR.pkl')
joblib.dump(value = [SVMC, ref_cols, target], filename = 'SVMC.pkl')
joblib.dump(value = [SVMC1, ref_cols, target], filename = 'SVMC1.pkl')
joblib.dump(value = [KNNC, ref_cols, target], filename = 'KNNC.pkl')
joblib.dump(value = [RFC, ref_cols, target], filename = 'RFC.pkl')
```

```python
def voting(dataF):
    X_ = dataF[ref_cols]

    predicts = []
    models = [LR, SVMC, SVMC1, RFC, KNNC]

    for item in models:
        predicts.append(item.predict(X_)[0])
    
    mean = (predicts[0]+predicts[1]+predicts[2]+predicts[3]+predicts[4])/5

    if mean > 0.65:
        return 1
    else:
        return 0

```

```python
ejemplo = df_.sample(1)
print('Class Real:', end=' ')
print(ejemplo['class'])
print('Class Predict:', end=' ')
print(voting(ejemplo))
```
